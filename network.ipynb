```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Sample threshold matrix
data = {
    'America': [1, 1, 0, 0, 1],
    'Canada': [0, 1, 1, 0, 0],
    'India': [1, 0, 1, 1, 1],
    'Gold': [0, 0, 0, 1, 1],
    'Copper': [1, 1, 1, 0, 1],
}

threshold_matrix = pd.DataFrame(data, index=['America', 'Canada', 'India', 'Gold', 'Copper'])

# Create a directed graph from the threshold matrix
G = nx.DiGraph()
for col in threshold_matrix.columns:
    for row in threshold_matrix.index:
        if threshold_matrix.loc[row, col] == 1:
            G.add_edge(row,col )

# Draw the graph
pos = nx.spring_layout(G)  # You can use different layout algorithms
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8, edge_color='gray')

# Show the plot
plt.show()
```

```python
avg_shortest_path_length = nx.average_shortest_path_length(G)

# Display the result
print(f"Average Shortest Path Length: {avg_shortest_path_length}")
```

```python
# Calculate node betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Display the result
print("Node Betweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"{node}: {centrality}")
```

```python
# Calculate directed node betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G, normalized=False)

# Display the result
print("Directed Node Betweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"{node}: {centrality}")
```

```python
# Calculate network density
num_nodes = len(G.nodes)
num_edges = len(G.edges)
density = num_edges / (num_nodes * (num_nodes - 1))

# Display the result
print(f"Network Density: {density}")
```

```python
# Calculate average clustering coefficient
average_clustering_coefficient = nx.average_clustering(G)

# Display the result
print(f"Average Clustering Coefficient: {average_clustering_coefficient}")
```

```python
# Calculate clustering coefficient for each node
clustering_coefficients = nx.clustering(G)

# Display the result
print("Clustering Coefficients for Each Node:")
for node, clustering_coefficient in clustering_coefficients.items():
    print(f"{node}: {clustering_coefficient}")
```
