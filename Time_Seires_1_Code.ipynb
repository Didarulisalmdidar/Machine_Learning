```python
x=dataset1.columns
```

```python
for col_name in x:
    if dataset1[col_name].isnull().sum() > 0:
        fit=dataset1[col_name].mean()
        dataset1[col_name].fillna(fit,inplace=True)
```

```python
dataset1.isnull().sum()
```

```python
dataset1.to_csv('dataset1.csv')
```

```python
dataset1.head()
```

```python
for col_name in x:
    print(col_name+'_Return')
```

```python
for col_name in x:
    print(col_name)
```

```python
dataset1_return=pd.DataFrame()
```

```python
dataset1_return
```

```python
for col_name in x:
    dataset1_return[col_name]=np.log(dataset1[col_name])-(np.log(dataset1[col_name]).shift(1))
```

```python
dataset1_return
```

```python
dataset1_return=dataset1_return.dropna()
```

```python
dataset1_return
```

```python
from sklearn.ensemble import RandomForestRegressor
feature_importance_scores = {}
for feature_name in dataset1_return.columns:
    y=dataset1_return[feature_name]
    y=y.iloc[1:]
    x=dataset1_return.drop(dataset1_return.index[-1])
    model=RandomForestRegressor()
    model.fit(x,y)
    score=model.feature_importances_
    feature_importance_scores[feature_name]=score
    print(feature_name)
    print(score)
    print(score.sum())
    
```

```python
importance_matrix = []
for feature_name, importance_score in feature_importance_scores.items():
    importance_matrix.append(importance_score)
importance_matrix = np.array(importance_matrix)
```

```python
feature_names = list(feature_importance_scores.keys())
importance_df = pd.DataFrame(importance_matrix, columns=feature_names, index=feature_names)
print(importance_df)
```
