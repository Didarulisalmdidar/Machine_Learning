```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
z=np.cos(x)
plt.plot(x,z,'r+')
#plt.show()
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'r.')
plt.plot(x,np.cos(x),'g.')
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
#bar graph
fig,ax=plt.subplots()
city=['Dhaka','Feni','nk']
people=[80,50,70]
ax.bar(city,people)
plt.xlabel('city')
plt.ylabel('people')
plt.title('people per city ')
```

```python
fig,ax=plt.subplots()
city=['Dhaka','Feni','nk']
people=[80,50,70]
ax.pie(people,labels=city,autopct='%1.1f%%')
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
#plt.plot(x,y)
```

```python
#2d scatter plot
fig=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Reds')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes([0,0,1,1])
dept=['cste','ice','eee','swe']
student=[400,100,240,300]
ax.bar(dept,student)
```

```python
fig=plt.figure()
ax=plt.axes([0,0,1,1])
dept=['cste','ice','eee','swe']
student=[400,100,240,300]
ax.pie(student,labels=dept,autopct='%1.1f%%')
```
